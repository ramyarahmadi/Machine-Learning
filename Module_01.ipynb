## 📖 1. Introduction to the ML Ecosystem

Before writing a single line of code, we must understand where Machine Learning fits in the broader landscape of Artificial Intelligence.

1. **Artificial Intelligence (AI):** The broad concept of machines being able to carry out tasks in a way that we would consider "smart."

A professional ML engineer follows a rigorous cycle:
**Problem Definition** $\rightarrow$ **Data Collection** $\rightarrow$ **Data Cleaning** $\rightarrow$ **EDA** $\rightarrow$ **Feature Engineering** $\rightarrow$ **Model Training** $\rightarrow$ **Evaluation** $\rightarrow$ **Deployment**.

## 🛠 2. Data Engineering: Handling the "Dirty" Data

In the real world, data is never clean. It is often messy, missing, and inconsistent. 
In this section, we will simulate a "dirty" dataset and apply professional cleaning techniques.

**The 5Vs Context:** As data grows in *Variety* and *Veracity*, our ability to clean and normalize it becomes the most critical step in the pipeline.


```python
# Creating a synthetic "dirty" dataset
data = {
    'Age': [25, 30, np.nan, 35, 40, 22, np.nan, 45, 50, 28],
    'Salary': [50000, 60000, 55000, np.nan, 80000, 45000, 52000, np.nan, 90000, 62000],
    'City': ['Tehran', 'Shiraz', 'Tehran', 'Isfahan', 'Shiraz', np.nan, 'Tehran', 'Isfahan', 'Tehran', 'Shiraz'],
    'Purchased': [0, 1, 0, 1, 1, 0, 1, 0, 1, 0]
}

df = pd.DataFrame(data)
print("Original Dirty Dataset:")
print(df)
print("\nMissing Values per Column:")
print(df.isnull().sum())
```

## 🧪 2.1 Handling Missing Values (Imputation)

We have two main strategies for missing data:
1. **Numerical Imputation:** Replacing missing numbers with the `Mean`, `Median`, or `Mode`.
2. **Categorical Imputation:** Replacing missing categories with the most frequent value (Mode).


```python
# 1. Handling Missing Numerical Values (using Median for Age and Salary)
imputer_num = SimpleImputer(strategy='median')
df[['Age', 'Salary']] = imputer_num.fit_transform(df[['Age', 'Salary']])

# 2. Handling Missing Categorical Values (using Mode for City)
imputer_cat = SimpleImputer(strategy='most_frequent')
df[['City']] = imputer_cat.fit_transform(df[['City']])

print("Cleaned Dataset after Imputation:")
print(df)
```

## ⚖️ 2.2 Feature Scaling (Standardization vs. Normalization)

Machine Learning algorithms (like KNN or SVM) are sensitive to the scale of data. 
- **Standardization (Z-score):** Rescales data to have a mean of 0 and a standard deviation of 1.
- **Normalization (Min-Max):** Rescales data to a fixed range, usually $[0, 1]$.

We will use **Standardization** here.


```python
# 3. Encoding Categorical Data (One-Hot Encoding for 'City')
df = pd.get_dummies(df, columns=['City'], drop_first=True)

# 4. Feature Scaling (Standardization for Age and Salary)
scaler = StandardScaler()
df[['Age', 'Salary']] = scaler.fit_transform(df[['Age', 'Salary']])

print("Final Processed Dataset (Ready for ML):")
print(df)
```

## 📊 2.3 Exploratory Data Analysis (EDA)

Visualizing the distribution of our scaled features helps us understand the data's spread.


```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['Age'], kde=True, color='skyblue')
plt.title('Distribution of Scaled Age')

plt.subplot(1, 2, 2)
sns.histplot(df['Salary'], kde=True, color='salmon')
plt.title('Distribution of Scaled Salary')

plt.tight_layout()
plt.show()
```
