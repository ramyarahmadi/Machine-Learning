# Explicit or non Explicit programming:

**Unlike explicit programming, where rules are manually defined by developers, our approach leverages implicit programming through Machine Learning. Instead of hard-coding specific relationships, the model learns the underlying patterns and parameters directly from the data.**

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

**Supervised Learning**

**In supervised learning, we need matching lengths for the X data and the corresponding y labels.**

“When they are not the same or have missing values, you need to fill them with None or another approach.And in principle, we are dealing with unsupervised learning.”

```python
X = np.array([100,150,180,200,250,300,320,350,400,450]) # area
y = np.array([10,15,18,20,21,31,33,36,39,42])           # price
```

```python
plt.xlim(0,600)
plt.ylim(0,50)
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(X,y)
```

**We expect the model to learn the relationship y=ax+b, which allows us to predict prices for new areas.**

The learning process starts by initializing the **model parameters**, a and b, to zero. To control the optimization process, we define a **hyperparameter** called max_iter, which specifies the maximum number of iterations the algorithm is allowed to perform. Through this iterative procedure, **the model continuously updates the parameters a and b to minimize the loss function and converge toward their optimal values**

```python
max_iter=100
a=0
b=0
```

```python
y_pred = a * X + b
```

```python
plt.scatter(X,y,c='g',label='y_true')
plt.plot(X,y_pred,'r--D', label="y'_prediction")
plt.xlim(-10,600)
plt.ylim(-10,50)
plt.xlabel('Area')
plt.ylabel('Price')
plt.legend()
```

**To evaluate the model’s performance**, we calculate the cost function by measuring the distance between the actual labels (y) and the predicted labels (y'). By using the **Mean Squared Error (MSE)**, we ensure that all residuals are treated as positive values, preventing the cancellation of errors. This function acts as the objective for our optimization process, guiding the model to minimize the total error.

```python
sum_error = sum(y_pred - y)
print(sum_error)
```

When initializing parameters (e.g., y=10 and a=0.01), **simply summing the errors can lead to incorrect results due to the cancellation of positive and negative values.** To accurately quantify the discrepancy between the predicted value (y') and the actual target (y), we must employ a loss function, such as Mean Squared Error (MSE), which squares the residuals to ensure all errors are positive and penalizes larger deviations more heavily.


```python
a=0.01
b=25
y_pred = a * X + b
```

```python
plt.scatter(X,y,c='g',label='y_true')
plt.plot(X,y_pred,'r--D', label="y'_prediction")
plt.xlim(-10,600)
plt.ylim(-10,50)
plt.xlabel('Area')
plt.ylabel('Price')
plt.legend()
```

**SUM ABSLUTE ERROR**

```python
error = y_pred - y
sum_error=np.sum(np.abs(error))
print(sum_error)
```

## Cost Function Implementation

To find the optimal values for aa and bb, we iterate through a predefined range of values using nested loops. For each combination of aa and bb, we evaluate the cost function and track the pair that results in the minimum error.

```python
for a in np.arange(-1,1,0.5):
    for b in np.arange(-50,51,5):
        y_pred = a * X + b
        plt.plot(X,y_pred,"r--D")
        sum_error=np.sum(np.abs(y_pred - y))
        print(a,b,sum_error)
plt.scatter(X,y,c="g")

```

For each candidate line (defined by a and b), we calculate the error using the cost function. The combination that yields the minimum error is identified as our **learned model.**
In this approach, we optimize the model by systematically searching through the parameter space to identify the configuration that **minimizes the cost function**. For example; when **a=0.0 ,b=30 ,error=97** is the best result.


```python
cost_function = []

for a in np.arange(-1,1,0.5):
    for b in np.arange(-50,51,5):
        y_pred = a * X + b
        #plt.plot(X,y_pred,"r--D")
        sum_errors=np.sum(np.abs(y_pred - y))
        print(a,b,sum_errors)
        cost_function.append(sum_errors)
#plt.scatter(X,y,c="g")
plt.plot(cost_function,'r')
```

**It is important to note that this result represents the optimal parameters within the defined search space only.**
